# Necesitamos hacer una sola request

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = f"https://housinganywhere.com/es/Espa%C3%B1a"
response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
grids = soup.find_all('a',attrs={'data-test-locator':'Listing Card'})
grids
```
