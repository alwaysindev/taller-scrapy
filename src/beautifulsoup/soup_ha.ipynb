# Necesitamos hacer una sola request

```python
import requests
from bs4 import BeautifulSoup

url = "https://housinganywhere.com/es/s/Barcelona--Espa%C3%B1a"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/114.0.0.0 Safari/537.36"
    ),
    "Accept-Language": "es-ES,es;q=0.9,en;q=0.8",
}

response = requests.get(url, headers=headers)

print(response.status_code, response.url)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
grids = soup.find_all('a',attrs={'data-test-locator':'ListingCard/Anchor'})
grids
```

## A veces necesitamos movernos para que carguen los datos JavaScript

```python
import re

url = f"https://housinganywhere.com/es/s/Barcelona--Espa%C3%B1a"
response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})

soup = BeautifulSoup(response.text, 'html.parser')


soup.find_all('a', attrs={'data-test-locator': re.compile('Listing\s*Card', re.I)})

print(len(soup.find_all('a', attrs={'data-test-locator': re.compile('Listing\s*Card', re.I)})))
# soup.find_all('a',attrs={'data-test-locator':'Listing Card'})
```
