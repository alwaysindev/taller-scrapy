```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Lista para guardar los datos
data = []

# Recorrer las 100 páginas
for i in range(1, 15):
    print(f"Página: {str(i)}")
    url = f"https://zacatrus.es/juegos-de-mesa.html?p={i}"
    response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    print(response)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrar todos los juegos en la página
        juegos = soup.find_all('div', class_='product-item-info')

        for juego in juegos:
            nombre = juego.find('strong', class_='product name product-item-name')
            precio = juego.find('span', class_='price')
            link = juego.find('a')
            imagen = juego.find('img')

            data.append([
                nombre.text.strip() if nombre else "N/A",
                precio.text.strip() if precio else "N/A",
                link.get('href') if link else "N/A",  # Ahora usa .get() en lugar de ['href']
                imagen.get('src') if imagen else "N/A"  # También usa .get() para evitar errores
            ])

# Crear DataFrame
df = pd.DataFrame(data, columns=["nombre", "precio", "link", "imagen"])

```
